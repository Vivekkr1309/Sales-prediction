```python
vk=sns.countplot(x ='Radio',data=df)
for bars in vk.containers:
    vk.bar_label(bars)
```

```python
vk=sns.countplot(x ='TV',data=df)
for bars in vk.containers:
    vk.bar_label(bars)
```

```python
vk=sns.countplot(x ='Newspaper',data=df)
for bars in vk.containers:
    vk.bar_label(bars)
```

```python
vk=sns.countplot(x ='Sales',data=df)
for bars in vk.containers:
    vk.bar_label(bars)
```

```python
sns.barplot(x='TV', y='Radio', data=df)
plt.title('Sales prediction of Tv and Radio')
plt.show()
```

```python
sns.barplot(x='Newspaper', y='Sales', data=df)
plt.title('Sales prediction of Sales and Newspaper')
plt.show()
```

```python
sns.pairplot(df[['TV', 'Newspaper', 'Radio', 'Sales']], hue='Sales')
plt.title('Pairwise Scatter Plot')
plt.show()
```

```python
data = pd.read_csv('advertising.csv')

print(data.head())
plt.figure(figsize=(8, 6))
plt.hist(data['TV'], bins=10, alpha=0.5, label='TV', edgecolor='black')
plt.hist(data['Radio'], bins=10, alpha=0.5, label='Radio', edgecolor='black')
plt.title('Sales Comparison: TV vs. Radio')
plt.xlabel('Amount Spent on TV/Radio')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
data = pd.read_csv('advertising.csv')

print(data.head())
plt.figure(figsize=(8, 6))
plt.hist(data['Newspaper'], bins=10, alpha=0.5, label='TV', edgecolor='black')
plt.hist(data['TV'], bins=10, alpha=0.5, label='Radio', edgecolor='black')
plt.title('Sales Comparison: Newspaper vs. TV')
plt.xlabel('Amount Spent on TV/Newspaper')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='TV', y='Radio', data=df)
plt.title('Sales based on TV and Radio')
plt.xlabel('TV')
plt.ylabel('Radio')
plt.show()
plt.figure(figsize=(6, 6))
sns.countplot(x='TV', hue='Newspaper', data=df)
plt.title('Sales based on TV and Newspaper')
plt.xlabel('TV')
plt.ylabel('Newspaper')
plt.show()
plt.figure(figsize=(8, 6))
sns.countplot(x='TV', hue='Sales', data=df)
plt.title('Sales based on TV and Sales')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```
